### Запис та зчитування з файлу, формат CSV - COMPLETED

```python
import numpy as np
import os

# Оновлений dtype з додаванням нового поля 'group'
dt = np.dtype([('name', np.str_, 20), ('mark', np.int8), ('group', np.str_, 10)])

# Початковий масив, додаємо нове поле 'group'
arr = [
    ("Bohdan", 4, "Group1"),
    ("Marta", 5, "Group2"),
    ("Noname", 0, "Group3"),
    ("Антон", 5, "КН-41")
]

# Створення масиву з заданим типом даних
C = np.array(arr, dtype=dt)

# Збереження масиву у файл .npz
np.savez("my_mass.npz", my_mass=C)
D = np.load("my_mass.npz")
print("Прочитаний з файлу: ", D["my_mass"])

# Збереження масиву у CSV файл
filename = "temp.csv"
np.savetxt(filename, C, fmt="%s,%d,%s", header="name,mark,group", delimiter=",")

# Перевірка чи файл створився
print(f"Перевіряємо чи файл {filename} створився: {os.listdir()}")

# Читання CSV файлу
print("Читаємо файл за допомогою оператора with та методу readlines")
with open(filename) as f:
    D = f.readlines()

print(f"Прочитаний файл:\n {D} \n- як бачимо - {len(D)} елементи є стрічками {type(D[0])}.")

# Зчитування значень з файлу у масив
D = np.genfromtxt(filename, dtype=dt, delimiter=",", names=True)
print(f"Зчитане значення з файла:\n {D}, як бачимо - {D.size} елементи це {type(D[0])} \n- вбудований клас бібліотеки numpy.")
print(f"Доступитись до певного елемента можна наступним чином, наприклад імя першого елементу: {D[0]['name']}")

# Читання вашого імені з масиву
your_name = D[D['name'] == "Антон"]['name']
print(f"Ваше ім'я, знайдене у масиві: {your_name[0]}")  # Вивести перший результат
```

### Множення матриць

```python
A = np.array([ [1, 2, 3], [2, 2, 2], [3, 3, 3] ])
B = np.array([ [3, 2, 1], [1, 2, 3], [-1, -2, -3] ])
print(f"Маємо два масиви:\n {A} \n {B}")

R = A * B
print(f"Тут виконується поелементне множення:\n {R}")

R = np.dot(A, B)
print(f"Тут, масиви перемножуються як матриці:\n {R}")

MA = np.asmatrix(A)
MB = np.asmatrix(B)

R = MA * MB
print(f"Тут масиви перетворені на матриці і далі перемножуються:\n {R}")
```

```python
import numpy as np

# Створення масиву
O = np.arange(6).reshape(3, 2)
O_r = O.reshape((2, 3))  # Зміна розмірності
O_f = O.flatten()  # Перетворення на одномірний масив
O_t = O.transpose()  # Транспонування масиву
O_fl = np.flip(O)  # Перевернути масив (по обох осях)
O_fa = np.flip(O, axis=1)  # Перевернути масив по стовпцях

# Виведення результатів
print(f"""{O}
Змінити розмірність shape {O.shape} на {O_r.shape} при цьому розмір size {O.size} = {O_r.size} та ndim {np.ndim(O)} = {np.ndim(O_r)}:
{O_r}
Також можна перетворити {np.ndim(O)}-х вимірний масив розмірністю {O.shape} на рядок:
{O_f} 
з розмірністю {O_f.shape} при цьому size = {O_f.size} та ndim стане {np.ndim(O_f)}-х вимірним 
Або транспонувати масив щоб він став {O_t.shape} але зверніть увагу на елементи:
{O_t}
Можна обернути масив (обертаємо по стовпцях і рядках одночасно): 
{O_fl}
Або обернути лише елементи в кожному рядку (або стовпці):
{O_fa}
""")

# Завдання 1:
homework(""" 
- Створіть вектор з 12 елементів, визначте його shape, size та ndim;
- Змініть розмірність перетворивши вектор у матрицю;
- Здійсніть транспонування отриманої матриці та оберніть її;
- Перетворіть отриманий багатовимірний масив назад у вектор (стрічку);
""")

# 1. Створення вектора з 12 елементів
vector = np.arange(12)
print(f"Вектор:\n{vector}")
print(f"Shape: {vector.shape}, Size: {vector.size}, Ndim: {vector.ndim}\n")

# 2. Зміна розмірності перетворивши вектор у матрицю
matrix = vector.reshape(3, 4)  # Перетворюємо вектор в матрицю 3x4
print(f"Матриця після reshape (3x4):\n{matrix}")
print(f"Shape: {matrix.shape}, Size: {matrix.size}, Ndim: {matrix.ndim}\n")

# 3. Транспонування отриманої матриці
transposed_matrix = matrix.transpose()  # Транспонування матриці
print(f"Транспонована матриця (4x3):\n{transposed_matrix}")
print(f"Shape: {transposed_matrix.shape}, Size: {transposed_matrix.size}, Ndim: {transposed_matrix.ndim}\n")

# 4. Обертання матриці
flipped_matrix = np.flip(transposed_matrix)  # Перевертаємо матрицю (по обох осях)
print(f"Перевернута матриця:\n{flipped_matrix}")
print(f"Shape: {flipped_matrix.shape}, Size: {flipped_matrix.size}, Ndim: {flipped_matrix.ndim}\n")

# 5. Перетворення багатовимірного масиву назад у вектор
flattened_vector = flipped_matrix.flatten()  # Перетворюємо матрицю назад у вектор
print(f"Вектор після перетворення з матриці:\n{flattened_vector}")
print(f"Shape: {flattened_vector.shape}, Size: {flattened_vector.size}, Ndim: {flattened_vector.ndim}\n")
```
